This project focuses on analyzing finance-based customer transactions using Python and Pandas.

Key tasks:

Cleaning and organizing data

Identifying high-value customers
Detecting high-debit/high-risk customers

Credit vs Debit distribution?

Which customers need attention (high debit)?

Summary report for Manager.

```python
#Data Setup
import pandas as pd

data1 = {
    "Customer": ["Rahul", "Rahul", "Anita", "John", "Anita", "Kiran", "John", "Rahul", "Kiran"],
    "Type": ["Credit", "Debit", "Credit", "Debit", "Debit", "Credit", "Debit", "Credit", "Debit"],
    "Amount": [5000, 2000, 8000, 1000, 3000, 9000, 2000, 4000, 3000]
}

df = pd.DataFrame(data1)
print(df)
```

```python
#Total trnasaction amount
print("Total Transaction Amount :", df["Amount"].sum())
```

```python
#Credit vs Debit Total
print("Total Credit vs Debit Amount :", df.groupby("Type")["Amount"].sum())
```

Is company more in credit or debit?
Accouding to the data, the company is more on credit from its customers

```python
#Customer total
maxi = df.groupby("Customer")["Amount"].sum().sort_values(ascending=False)
print(maxi)
```

According to the data-Customer-wise Total LIKE SQL GROUP BY, Customer "Kiran" has the highest total amount including both credit and debit and will be the Top Customer/High value customer.

```python
#Customer with high debit
debit = df[df["Type"]=="Debit"].groupby("Customer")["Amount"].sum().sort_values(ascending=False)
print(debit)
```

According to the data above, these are the customers with high debit and these customers need more attention / warnings.

Classify customer based on total spending:
VIP: > 10,000
Normal: 5,000 - 10,000
Low: < 5,000


```python
#Classifying customer
valued_Cust = df.groupby("Customer")["Amount"].sum().reset_index()
for index, row in valued_Cust.iterrows():
  if row["Amount"] > 10000:
    print(row["Customer"], "is a VIP Customer")
  elif row["Amount"] >= 5000:
    print(row["Customer"], "is a Normal customer")
  else:
    print(row["Customer"], "is a Low value Customer")
```

```python
#EXPORT LIKE EXCEL
#Exporting to excel
valued_Cust.to_excel("Debit_Customer_classification.xlsx", index=False)
```
